```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
```

```python
heart = pd.read_csv("heart.csv")
```

```python
heart.head() #loads head of data
```

```python
heart.info()
```

```python
sns.barplot(heart['sex'], heart['chol']) #plots bargraph
```

```python
sns.distplot(heart['chol']) # plots distriubution
```

```python
sns.jointplot(heart['thalachh'],heart['chol']) #bivariate function - visualising data in 2D
```

```python
sns.jointplot(heart['thalachh'],heart['chol'], kind="hex")
```

```python
sns.jointplot(heart['thalachh'],heart['chol'], kind="kde") #kernal density estimate
```

```python
sns.distplot (heart['thalachh'], kde=False, rug=True) #kernal density estimate
```

```python
heart.head()
```

```python
sns.pairplot(heart[['thalachh', 'chol','trtbps']]) #shows pairwise relationships in plot 
```

```python
sns.stripplot(heart['cp'], heart['chol'])  #categorical sorting of data
```

```python
sns.stripplot(heart['cp'], heart['chol'], jitter = True) #aims to get rid of overlaps?
```

```python
sns.boxplot(heart['cp'], heart['chol'], hue=heart['sex']) #plots boxplots, hue is differentiating
```

```python
sns.barplot(heart['cp'], heart['chol'], hue=heart['sex']) #barplot function operates on full dataset and shows an arbitrary estimate using the mean by default. when there are multiple observations in each category, it also uses bootstrapping to compute a comfidence interval after the estimate and plots that using error bars
```

```python
sns.countplot(heart['cp']) #counts value of single column?
```

```python
sns.pointplot(heart['cp'], heart['chol'], hue=heart['sex']) #makes relationship between more than one column - the relationship between chol and cp
```

```python
sns.lmplot(x="trtbps", y="thalachh", data=heart) #used to linear regression, plots the linear relationshio between two variables: trtbps and thalachh
```

```python
sns.lmplot(x="trtbps", y="thalachh", hue="sex", data=heart) #another type of lm plot (regression plot) with sex hue - the regression of each sex
```
